```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv(r"C:\Users\ads\Desktop\Week 1\Data\raw_analyst_ratings.csv")

# Show first few rows
df.head()

# Headline length
df['headline_length'] = df['headline'].str.len()
print("Average headline length:", df['headline_length'].mean())

# Count of articles per publisher
df['publisher'].value_counts().head(10).plot(kind='bar', title='Top Publishers')
plt.show()

# Article count by date
df['date'] = pd.to_datetime(df['date'])
df['date'].dt.date.value_counts().sort_index().plot(figsize=(10,4), title='Articles Over Time')
plt.xticks(rotation=45)
plt.show()
```
